```python

n = 100

for dimension in np.arange(2, 100):

    solution_buffer = 0

    for i in range(n):

        vecteur_random = np.random.rand(dimension, 1)
        matrice_random = np.random.rand(dimension, dimension)

        solution = np.linalg.solve(matrice_random, vecteur_random).round(0)

        if solution.sum() != 0:
            solution_buffer += 1


    print(f"Pour la dimension {dimension} le pourcentage de solution est de {round(solution_buffer/n*100,2)} %")
```
